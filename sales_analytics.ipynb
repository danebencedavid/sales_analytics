```python
import numpy as np
import matplotlib as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('sales.csv')
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
# Adding a new column as Hour

df['Hour'] = (df['Time']).dt.hour
```

## Visualization

```python
sns.set_theme()
ax = sns.countplot(x='Gender', data=df);
for label in ax.containers:
    ax.bar_label(label)
ax.set_title('Gender of Customers');
```

```python
ax = sns.countplot(x='Customer type', data=df);
for label in ax.containers:
    ax.bar_label(label)
ax.set_title('Type of customers');
```

```python
sns.histplot(data=df,x='Rating',kde=True,bins=15).set_title('Ratings of stores by Customers');
```

```python
sns.lineplot(data=df,x='Hour',y='Quantity').set_title('Product sold per Hour');
```

```python
ax = sns.countplot(data=df,y='Product line',width=0.45,order = df['Product line'].value_counts().index);
for label in ax.containers:
    ax.bar_label(label)
```

```python
sns.catplot(x='Product line',y='Unit price',hue='Gender',data=df,aspect= 2.2);
```

```python
ax = sns.countplot(data=df,x='Payment');
for label in ax.containers:
    ax.bar_label(label)
```

```python
sns.boxplot(data=df,x='Payment',y='Total',hue='Gender',showmeans=True);
```
